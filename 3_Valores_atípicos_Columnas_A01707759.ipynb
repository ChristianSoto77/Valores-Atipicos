Vamos a visualizar el histograma y el boxplot de todas las columnas cuantitativas del dataset a la vez

```python
#Realizamos diagrama de caja o bigote
data = df.drop(["Unnamed: 0","indice_tiempo"], axis = 1)

fig = plt.figure(figsize =(10, 6))
plt.boxplot(data,labels = data.columns)
plt.xticks(rotation = 90)
plt.show() #dibujamos el diagrama
```

# Aplicar métodos (Desviación Estándar y Rango Intercuartílico) para eliminar Outliers, comprobar con diagramas de caja por cada columna


```python
columnas = list(data.columns)
indices = [1,2,4,8,9,12,13,15,19,21,22,23] #Estas son las columnas que no tienen outliers
for i in range(len(indices)):
  indices[i] = indices[i]-1
indices
```

##**Variable ventas_precios_corrientes**

```python
#Realizamos histograma de variables cuantitativas
fig = plt.figure(figsize =(5, 3))
plt.hist(x=data[columnas[indices[0]]], color='blue', rwidth=0.50)
plt.title('Histograma de Ventas_Precio_Corrientes')
plt.ylabel('Frecuencia')
plt.show() #dibujamos el histograma
```

**MÉTODO DE IQR (RANGO INTER QUARTILICO)**



```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=data[columnas[indices[0]]]

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos Ouliers
outliers_iqr= data[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtenemos datos limpios
data_clean_iqr1= data[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data_clean_iqr1
```

```python
#Realizamos diagrama de caja o bigote
fig = plt.figure(figsize =(5, 3))
plt.boxplot(data_clean_iqr1[columnas[indices[0]]]) 
plt.title("Ventas precio corriente sin outliers")
plt.show() #dibujamos el diagrama
```
